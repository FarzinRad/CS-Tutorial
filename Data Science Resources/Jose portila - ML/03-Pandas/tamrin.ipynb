```python
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
#Wow! Vectorization is much faster! Keep np.vectorize() in mind for the future. 'jose portila'
#np.vectorize() or other work like that is not an efficent action and it's better use SQL instead.

df['tip_quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df
```

```python
df['tip_quality'] = df['tip_quality'].replace(to_replace='Other',value='geda')
df
```
